While your data companions rushed off to jobs in finance and government, you remained adamant that science was the way
for you. Staying true to your mission, you've since joined Pymaceuticals Inc., a burgeoning pharmaceutical company based 
out of San Diego, CA. Pymaceuticals specializes in drug-based, anti-cancer pharmaceuticals. In their most recent efforts,
they've since begun screening for potential treatments to squamous cell carcinoma (SCC), a commonly occurring form of skin
cancer.


As their Chief Data Analyst, you've been given access to the complete data from their most recent animal study.
In this study, 250 mice were treated through a variety of drug regimes over the course of 45 days. 
Their physiological responses were then monitored over the course of that time. Your objective is to analyze the data to 
show how four treatments (Capomulin, Infubinol, Ketapril, and Placebo) compare.

Results:

After a statistical analysis of the animal study for three treatments: Capomulin, Infubinol, Ketapril, and Placebo, the results show, Capolulin has the best performance in "Tumor & Metastatic Response to Treatment", and "Survival Rates During Treatment". Among the drugs studied, Capolulin is the most effective treatment as it shows higher survival rate.

```python
# Dependencies and Setup

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats 
# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
# File to Load 
clinical_trial = "data/clinicaltrial_data.csv"
mouse_drug= "data/mouse_drug_data.csv"
# Read the Mouse and Drug Data and the Clinical Trial Data
clinical_trial = pd.read_csv("data/clinicaltrial_data.csv")
mouse_drug = pd.read_csv("data/mouse_drug_data.csv")
# Combine the data into a single dataset
mouse_clinical = pd.merge(clinical_trial, mouse_drug,how='outer', on="Mouse ID")
# Display the data table for preview
mouse_clinical.head()
```

```python
mouse_clinical.describe()
```

## Tumor Response to Treatment

```python
# Store the Mean "Tumor Volume Data" Grouped by Drug and Timepoint 
group_mean = mouse_clinical.groupby(["Drug","Timepoint"]).mean()
group_mean.reset_index(level = None, inplace = True)
# Convert to DataFrame
tr_mean = pd.DataFrame(group_mean)
# Preview DataFrame 
tr_mean.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
group_stdem = mouse_clinical.groupby(["Drug","Timepoint"]).sem()
group_stdem.reset_index(level = None, inplace = True)
# Convert to DataFrame
tr_stdem_tumorvolm = pd.DataFrame(group_stdem)
tr_stdem_tumorvolm = tr_stdem_tumorvolm.drop(columns=["Mouse ID", "Metastatic Sites"])
# Preview DataFrame 
tr_stdem_tumorvolm.head()
```

```python
#Syntax: DataFrame.pivot(index=None, columns=None, values=None)[source]
# Minor Data Munging to Re-Format the Data Frames
trt = tr_mean.pivot(index ="Timepoint", columns = 'Drug', values = "Tumor Volume (mm3)")
trt.reset_index(level = None, inplace = True)
#Preview the Reformatting worked
trt
```

```python
table_fourdrugs_trt = trt[["Timepoint", "Capomulin", "Infubinol", "Ketapril", "Placebo"]]
table_fourdrugs_trt
```

```python
# Generate the Plot (with Error Bars)
x_axis = np.arange(0,50,5)
x_axis
```

```python
 #Generate the Plot (with Error Bars)
#https://matplotlib.org/3.1.1/gallery/lines_bars_and_markers/errorbar_limits_simple.html#sphx-glr-gallery-lines-bars-and-markers-errorbar-limits-simple-py
plt.figure(figsize = (10, 5))
plt.errorbar(x=table_fourdrugs_trt["Timepoint"],y=table_fourdrugs_trt["Capomulin"], yerr=None, linestyle=":", fmt="o", label="Capomulin", color="tab:red")
plt.errorbar(x=table_fourdrugs_trt["Timepoint"],y=table_fourdrugs_trt["Infubinol"], yerr=None, linestyle=":", fmt="v", label="Infubinol", color="tab:blue")
plt.errorbar(x=table_fourdrugs_trt["Timepoint"],y=table_fourdrugs_trt["Ketapril"], yerr=None, linestyle=":", fmt="s", label="Ketapril", color="tab:green")
plt.errorbar(x=table_fourdrugs_trt["Timepoint"],y=table_fourdrugs_trt["Placebo"], yerr=None, linestyle="-.", fmt="d", label="Placebo", color="tab:gray")
plt.ylabel("Tumor Volume(mm3)",fontsize="16")
plt.xlabel("Time (Days)",fontsize="16")
plt.title("Tumor Response to Treatment",fontsize="20")
plt.grid()
plt.legend(loc="center")
plt.legend()
plt.xlim(min(x_axis+3), max(x_axis)+3)
plt.show()
# Save the Figure
plt.savefig("Analysis/Fig1.png")
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
group_mean = mouse_clinical.groupby(["Drug","Timepoint"]).mean()
group_mean.reset_index(level = None, inplace = True)
# Convert to DataFrame
mrt_mean= pd.DataFrame(group_mean[:])
mrt_mean = mrt_mean.drop(columns="Tumor Volume (mm3)")
# Preview DataFrame 
mrt_mean.head()
```

```python
# Store the standard error associated with met. Site Data Grouped by Drug and Timepoint 
group_met_sem = mouse_clinical.groupby(["Drug","Timepoint"]).sem()
# Convert to DataFrame
mrt_response_sem = pd.DataFrame(group_met_sem)
mrt_response_sem = mrt_response_sem.drop(columns=["Mouse ID", "Tumor Volume (mm3)"])
# Preview DataFrame 
mrt_response_sem.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mrt = mrt_mean.pivot(index ="Timepoint", columns = 'Drug', values = "Metastatic Sites")
mrt.reset_index(level = None, inplace = True)
#Preview the Reformatting worked
mrt.head()
```

```python
table_fourdrugs_mrt = mrt[["Timepoint", "Capomulin", "Infubinol", "Ketapril", "Placebo"]]
table_fourdrugs_mrt
```

```python
 #Generate the Plot (with Error Bars)
#https://matplotlib.org/3.1.1/gallery/lines_bars_and_markers/errorbar_limits_simple.html#sphx-glr-gallery-lines-bars-and-markers-errorbar-limits-simple-py
plt.figure(figsize = (10, 5))
plt.errorbar(x=table_fourdrugs_trt["Timepoint"],y=table_fourdrugs_mrt["Capomulin"], yerr=None, linestyle=":", fmt="o", label="Capomulin", color="tab:red")
plt.errorbar(x=table_fourdrugs_trt["Timepoint"],y=table_fourdrugs_mrt["Infubinol"], yerr=None, linestyle=":", fmt="v", label="Infubinol", color="tab:blue")
plt.errorbar(x=table_fourdrugs_trt["Timepoint"],y=table_fourdrugs_mrt["Ketapril"], yerr=None, linestyle=":", fmt="s", label="Ketapril", color="tab:green")
plt.errorbar(x=table_fourdrugs_trt["Timepoint"],y=table_fourdrugs_mrt["Placebo"], yerr=None, linestyle="-.", fmt="d", label="Placebo", color="tab:gray")
plt.ylabel("Metastatic Sites",fontsize="16")
plt.xlabel("Time (Days)",fontsize="16")
plt.title("Metastatic During Treatment",fontsize="20")
plt.grid()
plt.legend(loc="center")
plt.legend()
plt.xlim(min(x_axis), max(x_axis)+5)
plt.show()
# Save the Figure
plt.savefig("Analysis/Fig2.png")
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
group_count = mouse_clinical.groupby(["Drug","Timepoint"]).count()["Tumor Volume (mm3)"]
# Convert to DataFrame
mice_count = pd.DataFrame({"Mouse Count": group_count})
mice_count.reset_index().head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mice_count = mice_count.reset_index()
srm = mice_count.pivot(index ="Timepoint", columns = "Drug") ["Mouse Count"]
#Preview the Reformatting worked
srm
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Generate the Plot (Accounting for percentages)
plt.figure(figsize = (10, 5))
plt.plot(100*srm["Capomulin"]/25, linestyle="-", label="Capomulin", color="tab:red")
plt.plot(100*srm["Infubinol"]/25, linestyle="-", label="Infubinol", color="tab:blue")
plt.plot(100*srm["Ketapril"]/25, linestyle="-", label="Ketapril", color="tab:green")
plt.plot(100*srm["Placebo"]/25, linestyle=":", label="Placebo", color="tab:gray")
plt.ylabel("Survival Rate %",fontsize="16")
plt.xlabel("Time (Days)",fontsize="16")
plt.title("Survival Rates Durning Treatment",fontsize="20")
plt.grid()
plt.legend(loc="center")
plt.legend()
# Preview the Data Frame
plt.show()
# Save the Figure
plt.savefig("Analysis/Fig4.png")
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
tumor_perct_change = 100 * (trt.iloc[-1] - trt.iloc[0])/trt.iloc[0]
# Display the data to confirm
tumor_perct_change
```

```python
# Store all Relevant Percent Changes into a Tuple
percent_change =(tumor_perct_change["Capomulin"], tumor_perct_change["Infubinol"],tumor_perct_change["Ketapril"],tumor_perct_change["Placebo"])
percent_change
```

```python
x= ("Capomulin", "Infubinol", "Ketapril", "Placebo")

fig, ax = plt.subplots()
index_length = np.arange(len(percent_change))
pass_values = ax.bar(index_length[0], percent_change[0], color="green", width = 1)
failed_values = ax.bar(index_length[1:], percent_change[1: ], color="red", width = 1)

plt.xticks(index_length, x)
plt.grid()

plt.ylabel("Survival Rate %",fontsize="16")
plt.title("Survival Rates Durning Treatment",fontsize="18")

def autolabelFail(rects):
   for rect in rects:
       height = rect.get_height()
       ax.text(rect.get_x() + rect.get_width()/2., 3,
               '%d%%' % int(height),
               ha='center', va='bottom', color="white")
def autolabelPass(rects):
   for rect in rects:
       height = rect.get_height()
       ax.text(rect.get_x() + rect.get_width()/2., -8,
               '-%d%% ' % int(height),
               ha='center', va='bottom', color="white")

autolabelFail(failed_values)
autolabelPass(pass_values)
```
